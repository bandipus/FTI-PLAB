```python
# EXERCISE 2a: Implement a right cyclic shift

# Function UAB_right_shift.
# * Parameter n: Integer, number of places to shift (>= 0)
# * Parameter L: List, list with values to shift
# * Returns: shifted list
#
def UAB_right_shift(n, L):
        
    #### IMPLEMENTATION GOES HERE ####
    
    n = n % len(L)
    result = []
    first_element = len(L) - n
    for i in range(first_element, len(L)):
        result.append(L[i])
        
    for i in range(len(L) - n):
        result.append(L[i])
    
    return result    
    
    ##################################
```

```python
# EXERCISE 2b: Generate the two codes and check they are equivalent. Show length, dimension and minimum distance of the generated code

#### IMPLEMENTATION GOES HERE ####

v = [0, 0, 1, 1, 0, 1, 0]
vectors_shift = [v]
for i in range(len(v)):
    vectors_shift.append(UAB_right_shift(i,v))
G = matrix(GF(2), vectors_shift)
C = LinearCode(G)

Cyclic_C = codes.CyclicCode(1+x+x^3, 7)

print(C)
print(Cyclic_C)
print("Son iguals")

set_C = set(C.list())
set_Cyclic_C = set(Cyclic_C.list())
print(set_C)
print(set_Cyclic_C)

if set_C == set_Cyclic_C:
    print("Conjunts iguals")
else:
    print("No iguals")

length_C = C.length()
dimension_C = C.dimension()
minimum_distance_C = C.minimum_distance()
print("Longitud: ")
print(length_C)
print("Dimensio: ")
print(dimension_C)
print("Distancia Minima: ")
print(minimum_distance_C)

##################################
```

```python
# EXERCISE 7: Compute the generator matrix

# Function UAB_gen_matrix.
# * Parameter g: generator polynomial
# * Parameter n: Integer, length of the code
# * Returns: generator matrix
#

def UAB_gen_matrix(g, n):
    
    G = None
    
    #### IMPLEMENTATION GOES HERE ####

    coef_g = g.coefficients(sparse=False)
    n_k = g.degree()
    k = abs(n_k - n)
    for i in range(k - 1):
        coef_g.append(0)
    test_matrix = []
    test_matrix.append(coef_g)
    for i in range(k - 1):
        coef_g = UAB_right_shift(1, coef_g)
        test_matrix.append(coef_g)
    G = matrix(GF(2), test_matrix)   
    
    ##################################
    
    return G
```

```python
# EXERCISE 8: Compute the control matrix

# Function UAB_con_matrix.
# * Parameter g: generator polynomial
# * Parameter n: Integer, length of the code
# * Returns: control matrix
#

def UAB_con_matrix(g, n):
    
    H = None  
    
    #### IMPLEMENTATION GOES HERE ####
    
    aux = x^n + 1
    hx = aux/g
    hx_poly = Z2X(hx)
    k = hx_poly.degree()
    cols = n - k
    hx_coeff = hx_poly.coefficients(sparse=False)
    hx_coeff.reverse()
    hasterisco = Z2X(hx_coeff)
    zerosaafegir = n-hasterisco.degree()
    for i in range(0,zerosaafegir - 1):
        hx_coeff.append(0)
    aux = []
    aux.append(hx_coeff)
    for i in range(0,cols - 1):
        hx_coeff = UAB_right_shift(1,hx_coeff)
        aux.append(hx_coeff)
    H = matrix(GF(2), aux)
    
    ##################################
    
    return H
```
